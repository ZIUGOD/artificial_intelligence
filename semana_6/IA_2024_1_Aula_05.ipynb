```python
import pandas as pd
import io
from sklearn.model_selection import train_test_split
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score

# importar o bd
from google.colab import files

uploaded = files.upload()

# Lendo o conjunto de dados
dados = pd.read_csv(io.BytesIO(uploaded["wine_dataset_henri.csv"]))
dados.head()

# Separando os dados em features (x) e rótulos (y)
x = dados[["fixed_acidity", "volatile_acidity", "citric_acid", "residual_sugar", "chlorides", "free_sulfur_dioxide", "total_sulfur_dioxide", "density", "pH", "sulphates", "alcohol", "quality"]]
y = dados[["style"]]

# Dividindo o conjunto de dados em treinamento e teste (80/20)
treino_x, teste_x, treino_y, teste_y = train_test_split(x, y, test_size=0.2, random_state=42)

# Instanciando o modelo LinearSVC
modelo = LinearSVC()

# Treinando o modelo usando os dados de treinamento
modelo.fit(treino_x, treino_y.values.ravel())

# Fazendo previsões nos dados de teste
previsoes = modelo.predict(teste_x)

# Calculando a acurácia com a divisão do dataset
acuracia_com_divisao = accuracy_score(teste_y, previsoes)

print("A acurácia após a divisão do dataset é de: {:.2%}".format(acuracia_com_divisao))

# Treinando o modelo sem dividir o conjunto de dados
modelo = LinearSVC()

# Treinando o modelo usando todos os dados
modelo.fit(x, y.values.ravel())

# Fazendo previsões nos mesmos dados usados para treinamento
previsoes = modelo.predict(x)

# Calculando a acurácia sem a divisão do dataset
acuracia_sem_divisao = accuracy_score(y, previsoes)

print("A acurácia sem a divisão do dataset é de: {:.2%}".format(acuracia_sem_divisao))
```
